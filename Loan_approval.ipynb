```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns

df = pd.read_csv("loan.csv")
```

# Data Preprocessing 

```python
#categorical data means gender,employee_category 
#numeric data means age,salary etc
# boths null values and missing should be handeled differently
categorical_col = df.select_dtypes(include=["object"]).columns
numeric_col = df.select_dtypes(include=["number"]).columns

print(categorical_col)
print(numeric_col)
```

```python
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy="mean") # strategy = mean means filling the null numeric value with mean of the numeric data
df[numeric_col] = num_imputer.fit_transform(df[numeric_col])
```

```python
cate_impute = SimpleImputer(strategy="most_frequent")
df[categorical_col] = cate_impute.fit_transform(df[categorical_col])
```

# Feature Encoding 

```python
# Feature Encoding 
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

col = ["Employment_Status","Marital_Status","Loan_Purpose","Property_Area","Gender","Employer_Category"]
onehot = OneHotEncoder(drop="first",sparse_output=False,handle_unknown="ignore")
encoded_col = onehot.fit_transform(df[col])
en_df = pd.DataFrame(encoded_col,columns= onehot.get_feature_names_out(col) ,index=df.index)
df = pd.concat([df.drop(columns=col),en_df],axis=1)
df
```

```python
le = LabelEncoder()
df["Education_Level"] = le.fit_transform(df["Education_Level"])
df["Loan_Approved"] = le.fit_transform(df["Loan_Approved"])
df
```

# Feature Scaling 

```python
# Feature_Scaling

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler()

x = df.drop("Loan_Approved",axis=1)
y = df["Loan_Approved"]

x_train,x_test,y_train,y_test= train_test_split(

    x,
    y,
    test_size = 0.2,
    random_state = 42
)

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)


```

# Logistic Regression 

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score

model_logistic = LogisticRegression()
model_logistic.fit(x_train_scaled,y_train)
log_predict = model_logistic.predict(x_test_scaled)

print("Logistic Regression Evaluation : ")
print("Accuracy Score : ", accuracy_score(y_test,log_predict))
print("Precision Score : ", precision_score(y_test,log_predict))
print("Recall Score : ", recall_score(y_test,log_predict))
print("F1 Score : ", f1_score(y_test,log_predict))

```

# KNeighborsClassifiers 

```python
from sklearn.neighbors import KNeighborsClassifier

model_kn = KNeighborsClassifier(n_neighbors = 3)
model_kn.fit(x_train_scaled,y_train)
knn = model_kn.predict(x_test_scaled)

print("Kneighbors Classifier Evaluation : ")
print("Accuracy Score : ", accuracy_score(y_test,knn))
print("Precision Score : ", precision_score(y_test,knn))
print("Recall Score : ", recall_score(y_test,knn))
print("F1 Score : ", f1_score(y_test,knn))
```

# Naive_bayes 

```python
from sklearn.naive_bayes import GaussianNB 

model_naive = GaussianNB()
model_naive.fit(x_train_scaled,y_train)
naive = model_naive.predict(x_test_scaled)

print("Naive bayes Evaluation : ")
print("Accuracy Score : ", accuracy_score(y_test,naive))
print("Precision Score : ", precision_score(y_test,naive))
print("Recall Score : ", recall_score(y_test,naive))
print("F1 Score : ", f1_score(y_test,naive))
```
