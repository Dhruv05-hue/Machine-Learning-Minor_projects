```python
import torch
import torch.nn as nn
import torch.optim as optim 
import torchvision
from torchvision.datasets import CIFAR10
```

```python
from torch.utils.data import DataLoader
import torchvision.transforms as transforms 

transform = transforms.Compose([

    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
 
])

trainset = CIFAR10(root = "./data",train=True,download=True,transform = transform)
testset = CIFAR10(root = "./data",train=False,download=True,transform = transform)
```

```python
trainloader = DataLoader(trainset,batch_size=64,shuffle=True)
testloader = DataLoader(testset,batch_size=64)
```

```python
import torch
import torch.nn as nn

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.conv_layer = nn.Sequential(

            # 1st Convolution Layer
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            # 2nd Convolution Layer
            nn.Conv2d(32,64,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            # 3rd Convolution Layer
            nn.Conv2d(64,128,kernel_size=3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),
        )

        self.fc_layers = nn.Sequential(

            nn.Linear(4*4*128,256),
            nn.ReLU(),

            nn.Linear(256,10)
        )

    def forward(self,x):

        x = self.conv_layer(x)

        x = x.view(x.size(0),-1)

        x = self.fc_layers(x)

        return x
```

# Training CNN

```python
import torch.optim as optim

model = CNN()
criteria = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())
```

```python
epochs = 10
for epoch in range(epochs):
    epoch_training_loss = 0.0
    model.train()
    for image,labels in trainloader:
           optimizer.zero_grad()
           outputs = model.forward(image)
           loss = criteria(outputs,labels)
           loss.backward()
           optimizer.step()

           epoch_training_loss += loss
    print(f"epoch={epoch+1}/{epochs}  and loss = {epoch_training_loss/len(trainloader)}")
```

# Model Evaluation 

```python
correct_values = 0.0
total_values = 0

model.eval()
with torch.no_grad():
    for image,labels in testloader:
        output = model.forward(image)
        _, predicted = torch.max(output,1)
        correct_values += (predicted == labels).sum().item()
        total_values += labels.size(0)
print(f"Accuracy is {correct_values/total_values * 100}")
```
